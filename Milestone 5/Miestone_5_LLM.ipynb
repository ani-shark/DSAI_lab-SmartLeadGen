```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch, json, re

MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tok = AutoTokenizer.from_pretrained(MODEL_ID)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=torch.float16
)

def chat(messages, max_new_tokens=220, temperature=0.6, top_p=0.9, repetition_penalty=1.1):
    # build chat prompt
    text = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tok([text], return_tensors="pt").to(model.device)
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=temperature,
            top_p=top_p,
            repetition_penalty=repetition_penalty,
            eos_token_id=tok.eos_token_id,
        )
    # decode only the new tokens
    gen = out[0][inputs["input_ids"].shape[1]:]
    resp = tok.decode(gen, skip_special_tokens=True)
    return resp
```

```python
SYS_PROMPT = """
You write short, polite B2B cold emails.

Your job:
- Read the lead_profile and style_profile I give you.
- Then write a ready-to-send cold email for that specific person.

Hard rules:
- 60–120 words.
- One clear call to action.
- No links in the first email.
- Use ONLY information from lead_profile. Do not invent facts.
- Use greeting and closing from style_profile.
- DO NOT output instructions, guidelines, or placeholders.
- DO NOT explain what you are doing.

Output format (VERY IMPORTANT):
Return ONLY a single JSON object with this exact shape:
{"subject": "<short subject line>", "body": "<full email body text>"}
"""
```

```python
MBTI_TONE = {
 "ENTJ":{"assertiveness":"high","formality":"high","avg_wps":18,"cta_style":"decisive"},
 "INTJ":{"assertiveness":"med-high","formality":"high","avg_wps":17,"cta_style":"analytical"},
 "ENFJ":{"assertiveness":"medium","formality":"medium","avg_wps":20,"cta_style":"collaborative"},
 "ENFP":{"assertiveness":"medium","formality":"medium","avg_wps":21,"cta_style":"story"},
}
DEFAULT_TONE = {"assertiveness":"medium","formality":"medium","avg_wps":19,"cta_style":"consultative"}

def mbti_to_style_profile(mbti: str):
    base = MBTI_TONE.get((mbti or "").upper(), DEFAULT_TONE)
    return {
        "style_name": f"MBTI:{mbti or 'Consultative'}",
        "avg_words": 95,
        "avg_wps": base["avg_wps"],
        "formality": base["formality"],
        "assertiveness": base["assertiveness"],
        "cta_style": base["cta_style"],
        "greeting": "Hi {first_name},",
        "closing": "Best regards,\n{sender_name}",
        "quirks": []
    }

def normalize_profile(j):
    j = {**{"objective":"", "optional_context":"", "profile_data":{}}, **j}
    p = j["profile_data"]
    p.setdefault("name","")
    p.setdefault("headline","")
    p.setdefault("career_history",[])
    p.setdefault("skills",[])
    p.setdefault("recent_activity",[])
    p.setdefault("company_updates",[])
    p.setdefault("technical_contributions",[])
    p.setdefault("inferred_mbti","")
    return j

def build_lead_profile(j):
    p = j["profile_data"]
    role = p["career_history"][0]["role"] if p.get("career_history") else ""
    company = p["career_history"][0]["company"] if p.get("career_history") else ""
    return {
      "name": p.get("name",""),
      "headline": p.get("headline",""),
      "role": role,
      "company": company,
      "skills": p.get("skills", [])[:8],
      "recent_activity": p.get("recent_activity", [])[:3],
      "company_updates": p.get("company_updates", [])[:2],
      "technical_contributions": p.get("technical_contributions", [])[:3],
      "objective": j.get("objective",""),
      "optional_context": j.get("optional_context","")
    }

def generate_email(user_json, sender_name="Team", address="Your Address"):
    j = normalize_profile(user_json)
    lead_profile = build_lead_profile(j)
    mbti = j["profile_data"].get("inferred_mbti","")
    style_profile = mbti_to_style_profile(mbti)

    user_prompt = (
      "lead_profile = " + json.dumps(lead_profile, ensure_ascii=False) + "\n" +
      "style_profile = " + json.dumps(style_profile, ensure_ascii=False)
    )

    raw = chat(
        [
          {"role":"system", "content": SYS_PROMPT},
          {"role":"user", "content": user_prompt},
        ],
        max_new_tokens=220,
        temperature=0.6,
        top_p=0.9,
        repetition_penalty=1.1,
    )

    print("RAW MODEL OUTPUT:\n", raw, "\n--- END RAW ---")  # DEBUG

    # Try to parse JSON
    email = {"subject":"", "body":"", "raw": raw}
    try:
        # if it already looks like pure JSON
        txt = raw.strip()
        if txt.startswith("{") and txt.endswith("}"):
            parsed = json.loads(txt)
        else:
            # extract first {...}
            m = re.search(r'\{.*\}', txt, re.S)
            if m:
                parsed = json.loads(m.group(0))
            else:
                parsed = {}
        email["subject"] = parsed.get("subject","").strip()
        email["body"] = parsed.get("body","").strip()
    except Exception as e:
        print("JSON parse failed, falling back to raw text:", e)

    # Fallback: if body still empty, just use raw
    if not email["body"].strip():
        email["subject"] = email["subject"] or "Quick idea for you"
        email["body"] = raw.strip()

    # Add signature/compliance
    if email["body"] and "unsubscribe" not in email["body"].lower():
        email["body"] += f"\n\n—\n{address}\nIf this isn’t relevant, reply “unsubscribe” and we won’t contact you again."
    email["body"] = email["body"].replace("{sender_name}", sender_name)
    return email
```

```python
import re
import textwrap
import pandas as pd
import numpy as np

# -------------------------------------------------------------------
# 0) If not already defined, rebuild test_leads from your user_jsons
# -------------------------------------------------------------------
try:
    test_leads
except NameError:
    test_leads = [
        user_json,
        user_json_2,
        user_json_3,
        user_json_4,
        user_json_5,
        user_json_6,
        user_json_7,
    ]

SENDER_NAME = "Sumit"
ADDRESS = "123 OMR, Chennai"

# -------------------------------------------------------------------
# 1) Run your model on all leads
# -------------------------------------------------------------------
def run_model_on_leads(leads):
    results = []
    for i, lead in enumerate(leads, start=1):
        email = generate_email(lead, sender_name=SENDER_NAME, address=ADDRESS)

        # fallbacks just in case email is not exactly a dict
        subj = ""
        body = ""
        if isinstance(email, dict):
            subj = email.get("subject", "")
            body = email.get("body", "")
        elif isinstance(email, str):
            body = email

        results.append({
            "idx": i,
            "lead_name": lead["profile_data"]["name"],
            "lead_objective": lead["objective"],
            "email_subject": subj,
            "email_body": body,
            "lead_profile": lead["profile_data"],
        })
    return results

results = run_model_on_leads(test_leads)
print(f"Generated {len(results)} emails.")
print("Example subject/body for the first lead:\n")
print("SUBJECT:", results[0]["email_subject"])
print("BODY:\n", textwrap.fill(results[0]["email_body"], width=80))

# -------------------------------------------------------------------
# 2) Helper functions for rule-based evaluation
# -------------------------------------------------------------------
def word_count(text: str) -> int:
    tokens = re.findall(r"\w+", text)
    return len(tokens)

def has_greeting(body: str, name: str) -> bool:
    first_lines = body.strip().splitlines()[:2]
    joined = " ".join(first_lines).lower()
    name_tokens = name.lower().split()
    if not name_tokens:
        return False
    name_core = name_tokens[-1]  # last token (Iyer, Mehta, etc.)
    patterns = [
        f"hi {name_core}",
        f"hello {name_core}",
        f"dear {name_core}"
    ]
    return any(pat in joined for pat in patterns)

def has_closing(body: str, sender_name: str) -> bool:
    lines = [l.strip().lower() for l in body.strip().splitlines() if l.strip()]
    if not lines:
        return False
    last_lines = " ".join(lines[-3:])  # last 3 non-empty lines
    sender_core = sender_name.lower().split()[0]  # "sumit"
    closing_words = ["regards", "best", "thanks", "thank you", "cheers"]
    return (sender_core in last_lines) and any(cw in last_lines for cw in closing_words)

def has_links(body: str) -> bool:
    b = body.lower()
    return ("http://" in b) or ("https://" in b) or ("www." in b)

def estimate_cta_count(body: str) -> int:
    cta_verbs = [
        "schedule", "jump on", "connect", "chat", "call", "discuss",
        "explore", "walk you through", "take a look", "set up", "hop on",
        "run you through"
    ]
    sentences = re.split(r"[.!?]\s+", body.strip())
    count = 0
    for s in sentences:
        s_low = s.lower()
        if any(v in s_low for v in cta_verbs):
            count += 1
    return count

def uses_name_or_company(body: str, lead_profile: dict) -> bool:
    text = body.lower()
    # check name
    name = lead_profile.get("name", "")
    for token in name.lower().split():
        if token and token in text:
            return True
    # check company (from career_history)
    for ch in lead_profile.get("career_history", []):
        c = ch.get("company", "").lower()
        if c and c in text:
            return True
    return False

def uses_objective_keywords(body: str, objective: str) -> bool:
    stop = {"for", "and", "the", "to", "of", "in", "a", "an", "on", "at", "with"}
    obj_tokens = [t.lower() for t in re.findall(r"\w+", objective) if t.lower() not in stop]
    text = body.lower()
    return any(tok in text for tok in obj_tokens)

# -------------------------------------------------------------------
# 3) Build per-email metrics
# -------------------------------------------------------------------
rows = []
for r in results:
    subj = r["email_subject"] or ""
    body = r["email_body"] or ""
    profile = r["lead_profile"]
    objective = r["lead_objective"]

    subject_ok = isinstance(subj, str) and len(subj.strip()) > 0
    body_ok = isinstance(body, str) and len(body.strip()) > 0

    wc = word_count(body)
    wc_ok = 60 <= wc <= 120

    greeting_ok = has_greeting(body, profile["name"])
    closing_ok = has_closing(body, SENDER_NAME)
    links_present = has_links(body)

    cta_count = estimate_cta_count(body)
    single_cta_ok = (cta_count == 1)

    personalization_ok = uses_name_or_company(body, profile)
    objective_ref_ok = uses_objective_keywords(body, objective)

    rows.append({
        "idx": r["idx"],
        "lead_name": r["lead_name"],
        "subject_ok": subject_ok,
        "body_ok": body_ok,
        "word_count": wc,
        "word_count_ok_60_120": wc_ok,
        "greeting_ok": greeting_ok,
        "closing_ok": closing_ok,
        "links_present": links_present,
        "single_cta_ok": single_cta_ok,
        "cta_count": cta_count,
        "personalization_ok": personalization_ok,
        "objective_ref_ok": objective_ref_ok,
    })

metrics_df = pd.DataFrame(rows)
print("\nPer-email rule metrics:\n")
display(metrics_df)

# -------------------------------------------------------------------
# 4) Aggregate “model-level” metrics (for your Milestone 5 table)
# -------------------------------------------------------------------
agg = {
    "n_emails": len(metrics_df),
    "subject_valid_rate": metrics_df["subject_ok"].mean(),
    "body_valid_rate": metrics_df["body_ok"].mean(),
    "word_count_in_range_rate": metrics_df["word_count_ok_60_120"].mean(),
    "greeting_ok_rate": metrics_df["greeting_ok"].mean(),
    "closing_ok_rate": metrics_df["closing_ok"].mean(),
    "no_links_rate": (metrics_df["links_present"] == False).mean(),
    "single_cta_ok_rate": metrics_df["single_cta_ok"].mean(),
    "personalization_ok_rate": metrics_df["personalization_ok"].mean(),
    "objective_ref_ok_rate": metrics_df["objective_ref_ok"].mean(),
    "avg_word_count": metrics_df["word_count"].mean(),
    "min_word_count": metrics_df["word_count"].min(),
    "max_word_count": metrics_df["word_count"].max(),
}
agg_df = pd.DataFrame([agg]).T
agg_df.columns = ["value"]

print("\nAggregated metrics:\n")
display(agg_df)

# -------------------------------------------------------------------
# 5) Quality score (1–10) instead of "failed cases"
# -------------------------------------------------------------------
def compute_quality_score(row):
    """
    Turn rule-based checks into a 1–10 score.
    This is a simple, positive scoring scheme.
    """
    score = 0

    # 1 point each for basic structure
    if row["subject_ok"]:
        score += 1
    if row["body_ok"]:
        score += 1

    # 2 points if word count obeys 60–120 (important constraint)
    if row["word_count_ok_60_120"]:
        score += 2

    # Greeting & closing: 1 point each
    if row["greeting_ok"]:
        score += 1
    if row["closing_ok"]:
        score += 1

    # No links in first email: 1 point
    if not row["links_present"]:
        score += 1

    # Exactly one clear CTA: 1 point
    if row["single_cta_ok"]:
        score += 1

    # Personalization + objective usage: 1 point each
    if row["personalization_ok"]:
        score += 1
    if row["objective_ref_ok"]:
        score += 1

    # Ensure score is between 1 and 10 (never 0)
    score = max(1, min(10, score))
    return score

metrics_df["quality_score_1_to_10"] = metrics_df.apply(compute_quality_score, axis=1)

print("\nPer-email quality scores (1–10):\n")
display(metrics_df[["idx", "lead_name", "quality_score_1_to_10", "word_count"]])

avg_score = metrics_df["quality_score_1_to_10"].mean()
print(f"\nAverage quality score across all emails: {avg_score:.2f} / 10")

# -------------------------------------------------------------------
# 6) (Optional) Manual scores if you want to eye-ball & override
# -------------------------------------------------------------------
# Uncomment and fill this list if you want *your own* 1–10 ratings
# after reading each email.

# manual_scores = [5, 6, 4, 5, 4, 5, 6]  # REPLACE with your own ratings
# metrics_df["manual_score_1_to_10"] = manual_scores
# print("\nPer-email manual scores:\n")
# display(metrics_df[["idx", "lead_name", "manual_score_1_to_10"]])
# print("\nAverage manual score:", metrics_df["manual_score_1_to_10"].mean())
```
