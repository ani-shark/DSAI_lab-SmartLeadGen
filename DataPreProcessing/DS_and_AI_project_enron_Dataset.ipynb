```python

outreach_keywords = [
    'introduction', 'request', 'meeting', 'demo', 'call', 'connect',
    'follow-up', 'proposal', 'opportunity', 'partnership', 'pricing',
    'quote', 'product', 'service', 'solution', 'business'
]


df['subject'] = df['subject'].fillna('').astype(str)
df['body'] = df['body'].fillna('').astype(str)


keyword_pattern = '|'.join(outreach_keywords)


print("Filtering for emails containing outreach keywords...")
filtered_df = df[
    df['subject'].str.contains(keyword_pattern, case=False) |
    df['body'].str.contains(keyword_pattern, case=False)
].copy()

print(f"\nOriginal number of emails: {len(df)}")
print(f"Number of emails after filtering: {len(filtered_df)}")
print(f"\nFound {len(filtered_df)} potentially relevant emails.")


print("\n--- Examples of Filtered Emails ---")
for i, row in filtered_df.head(3).iterrows():
    print(f"Subject: {row['subject']}")
    print("---")

    print(f"{row['body'][:500]}...")
    print("\n=====================================\n")
```

### Cleaning the Email Body

```python

patterns_to_check = {
    "Forwarded/Replied": r'-----original message-----',
    "From Line": r'from:',
    "To Line": r'to:',
    "CC Line": r'cc:',
    "Sent Line": r'sent:',
    "Subject Line": r'subject:',
    "Signature": r'--'
}

print("Checking for the existence of cleaning patterns in the 'body' column...\n")

total_emails = len(filtered_df)

if total_emails > 0:
    for name, pattern in patterns_to_check.items():

        count = filtered_df['body'].str.contains(pattern, case=False, regex=True).sum()


        percentage = (count / total_emails) * 100

        print(f"- Pattern '{name}' ({pattern}): Found in {count} emails ({percentage:.2f}%)")
else:
    print("The filtered_df is empty. No patterns to check.")
```

```python
import re

def clean_email_body(text):

    text = re.sub(r'(?is)^-{10,}\s*forwarded by.*?-{10,}\s*$', '', text, flags=re.MULTILINE)
    text = re.sub(r'(?is)-----original message-----.*', '', text, flags=re.DOTALL)
    text = re.sub(r'--\s*\n.*', '', text, flags=re.DOTALL) # Remove signatures
    text = text.strip()

    lines = text.splitlines()
    body_start_index = 0
    header_pattern = re.compile(r'^[a-zA-Z0-9\-\._]+:\s*.*') # it matches "To:", "Subject:"

    for i, line in enumerate(lines):

        if line.strip() and not header_pattern.match(line):
            body_start_index = i
            break


    body = '\n'.join(lines[body_start_index:])

    return body.strip()


print("Applying the cleaning function...")
filtered_df['cleaned_body'] = filtered_df['body'].apply(clean_email_body)
print("Cleaning complete.")

print("Searching for a forwarded email to demonstrate the fix...")

example_df = filtered_df[filtered_df['body'].str.contains(r'forwarded by', case=False)]

if not example_df.empty:
    example_index = example_df.index[0]
    raw_body = filtered_df.loc[example_index, 'body']
    cleaned_body = filtered_df.loc[example_index, 'cleaned_body']

    print("\n BEFORE CLEANING: \n")
    print(raw_body)
    print("\n-------------------------\n")
    print(" AFTER CLEANING: \n")
    print(cleaned_body)
else:
    print("\nCould not find a clear example of a forwarded email in your dataset.")
```
